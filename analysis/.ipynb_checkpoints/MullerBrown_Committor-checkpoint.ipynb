```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import subprocess
import pandas as pd
import mlcolvar
import mlcolvar.utils.plot
from mlcolvar.data import DictDataset,DictModule
from mlcolvar.utils.plot import muller_brown_potential, plot_isolines_2D, plot_metrics
from mlcolvar.utils.io import load_dataframe,create_dataset_from_files

import os

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

import random

import plumed

torch.manual_seed(42)

#muller-brown potential

def V(x,y):
    prefactor = 0.15
    A=(-200,-100,-170,15)
    a=(-1,-1,-6.5,0.7)
    b=(0,0,11,0.6)
    c=(-10,-10,-6.5,0.7)
    x0=(1,0,-0.5,-1)
    y0=(0,0.5,1.5,1)
    offset = -146.7

    v = -prefactor*offset
    for i in range(4):
        v += prefactor * A[i]*np.exp( a[i]*(x-x0[i])**2 + b[i]*(x-x0[i])*(y-y0[i]) + c[i]*(y-y0[i])**2 )
    return v

MULLER_BROWN_FORMULA='0.15*(146.7-200*exp(-1*(x-1)^2+0*(x-1)*(y-0)-10*(y-0)^2)-100*exp(-1*(x-0)^2+0*(x-0)*(y-0.5)-10*(y-0.5)^2)-170*exp(-6.5*(x+0.5)^2+11*(x+0.5)*(y-1.5)-6.5*(y-1.5)^2)+15*exp(0.7*(x+1)^2+0.6*(x+1)*(y-1)+0.7*(y-1)^2))'

# Visualize
nx, ny = (100, 100)
xlims=(-1.8, 1.2)
ylims=(-0.4, 2.1)
x = np.linspace(xlims[0],xlims[1], nx)
y = np.linspace(ylims[0],ylims[1], ny)
xv, yv = np.meshgrid(x, y)

pot = V(xv,yv)
max_pot = 26
pot[pot>max_pot] = max_pot # np.NaN

fig,ax = plt.subplots(dpi=100)
pp = ax.contourf(xv,yv,pot,16,cmap='rainbow_r')
ax.set_aspect('equal')
plt.colorbar(pp,ax=ax)
plt.show()
```

```python
folder = '../raw_data/Muller_Brown/'
colvar=plumed.read_as_pandas(folder+"COLVAR_DeepLNE")
print(len(colvar))

# plot pot isolines + samples
_, ax = plt.subplots(1, figsize=(6,5), dpi=100)
ax.contour(xv,yv,pot,10,cmap='Greys_r')
ax.set_title(f'Muller-Brown potential')
colvar.plot.scatter('p.x','p.y',c='opes.bias',cmap='fessa',s=1,ax=ax)
plt.show()

colvar.plot('time','p.y')
plt.show()
```

```python
start = 0.56
end = 0.75
resolution = 5
bin_edges = np.linspace(start, end, resolution + 1)

all_starting_points=[]

for i in range(resolution):
    temp=[]
    for e,elem in enumerate(colvar['model.node-0'].to_numpy()):
        if bin_edges[i] <= elem <= bin_edges[i+1]:
            temp.append(colvar.iloc[e,1:3].to_numpy())
            
    while len(temp) < 300:
        temp.append(temp[random.randint(0, len(temp)-1)]+ (0.01 * np.random.uniform(-1, 1, size=(2,))))
    print(len(temp))
    all_starting_points.append(np.array(temp))


plt.figure()
plt.contour(xv,yv,pot,resolution,cmap='Greys_r')
for i in range(resolution):
    cmap = plt.get_cmap('rainbow_r')
    print(all_starting_points[i].shape)
    plt.scatter(all_starting_points[i][:,0],all_starting_points[i][:,1],color=cmap(i/resolution),s=1)
plt.show()
```

```python
for a,starting_points in enumerate(all_starting_points):
    print(a)
    for c,c_run in enumerate(starting_points):
        #print(c)
        x_start=c_run[0]
        y_start=c_run[1]

        with open(folder+"md_potential","w") as f:
            print("""#! FIELDS idx_dim1 idx_dim2 pot.coeffs index description
#! SET type LinearBasisSet
#! SET ndimensions  2
#! SET ncoeffs_total 9 
#! SET shape_dim1  3
#! SET shape_dim2  3
   0       0         0.0000000000000000e+00       0  1*1
#!-------------------
""",file=f)

# md_input for ves_md_linearexpansion
        with open(folder+"md_input","w") as f:
            print("""
nstep                   2000000
tstep                   0.005
temperature             1.0
friction                10.0
random_seed             1  
plumed_input            plumed.dat
dimension               2
replicas                1
basis_functions_1       BF_POWERS ORDER=2 MINIMUM=-4.0 MAXIMUM=+4.0
basis_functions_2       BF_POWERS ORDER=2 MINIMUM=-4.0 MAXIMUM=+4.0
input_coeffs            md_potential
initial_position        %s,%s
output_potential        out_potential.data
output_potential_grid   100
output_histogram        histogram.data
""" %(x_start,y_start),file=f)

        with open(folder+"plumed.dat","w") as f:
            print(f"""# vim:ft=plumed
UNITS NATURAL
p: POSITION ATOM=1
ene: CUSTOM ARG=p.x,p.y PERIODIC=NO FUNC={MULLER_BROWN_FORMULA}
pot: BIASVALUE ARG=ene

COMMITTOR ...
ARG=p.x,p.y
STRIDE=10
BASIN_LL1=0.5,-0.1
BASIN_UL1=0.7,0.1 
BASIN_LL2=-0.7,1.4
BASIN_UL2=-0.5,1.7 
...

lwall: LOWER_WALLS ARG=p.x KAPPA=1000 AT=-1.3
uwall: UPPER_WALLS ARG=p.x KAPPA=1000 AT=+1.0

PRINT STRIDE=20 ARG=* FILE=COLVAR_committor_%s_%s
""" %(str(a),str(c)),file=f)
        PLUMED_EXE="plumed"
        subprocess.run(f"{PLUMED_EXE} ves_md_linearexpansion < md_input", cwd=folder, shell=True)
        subprocess.run("rm out* histogram* *log potential-coeffs.out.data stats.out", cwd=folder, shell=True)
```

```python
def compute_committor(path_to_file,n_sim):

    stateA_counter=0
    stateB_counter=0
    for c in range(n_sim):

        colvar = load_dataframe(path_to_file+"%s" %str(c))
        last_position_x=colvar.iloc[-1,1]
        if last_position_x < 0:
            stateA_counter+=1
        else:
            stateB_counter+=1

    return stateA_counter/n_sim

folder = '../raw_data/Muller_Brown/'
    
bin_edges=np.array(bin_edges)

plt.figure()
committors=[]
for a,starting_points in enumerate(all_starting_points):
    c=compute_committor(folder+'COLVAR_committor_%s_' %a,len(starting_points))
    
    committors.append(c)

plt.plot(0.5*(bin_edges[:resolution]+bin_edges[1:resolution+1]),committors,'o-')
plt.show()

data=np.load(folder+"MB_FES.npy" )

plt.figure()

for bin_edge in bin_edges:
    plt.axvline(x=bin_edge, color='black', linestyle='-', linewidth=1)

plt.errorbar(data[:,0], data[:,1]*2.479, yerr=data[:,2]*2.479, color='#0077cc',linewidth=2)
plt.plot(data[:,0], data[:,1]*2.479, color='#0077cc',linewidth=4)
plt.ylabel('free energy in kJ/mol')
plt.xlabel('path cv')

ax2 = plt.gca().twinx()
ax2.plot(0.5*(bin_edges[:resolution]+bin_edges[1:resolution+1]),committors,'o-',color='#ff9900',markersize=10,linewidth=3)
ax2.set_ylabel('Committor', color='#ff9900',fontsize=22)
ax2.tick_params(axis='y', colors='#ff9900',labelsize=18)
plt.tight_layout()
plt.show()
```
